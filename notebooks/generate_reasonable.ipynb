```python
import pandas as pd
import numpy as np
import csv
import random

from faker import Faker
from time import time
  
```

```python
fake = Faker()
```

```python
# changing column data with faker

accounts_list = ['Savings Account', 'Current Account']
sex_list = ['Male', 'Female']
#nationality = ['Zimbabwe', 'Zimbabwe', 'Zimbabwe', 'Zimbabwe', 'Zimbabwe', 'Zimbabwe', 'South Africa', 'Botswana', 'Nigeria', 'Mozambique', 'United Kingdom', 'United States', 'China', 'India']
nationality = [239, 239, 239, 239, 239, 239, 193, 28, 156, 145, 222, 223, 44, 99]
# names_list = [fake.name_female(), fake.name_male()]

# Data Generator function
 
def fake_data():
    return{'Name': fake.name(), 
           'Gender': random.choice(sex_list),
           'Address': fake.street_address(), 
           'Nationality': random.choice(nationality), 
           'Account_Type': random.choice(accounts_list), 
           'Age': random.randint(0, 2), 
           'Education': random.randint(0, 1), 
           'Employment': random.randint(0, 2),
           'Salary': random.randint(0, 2),
           'Employer_Stability': random.randint(0, 1),
           'Customer_Loyalty': random.randint(0, 10),
           'Balance': random.randint(0, 2),
           'Residential_Status': random.randint(0, 1)
          }

t0 = time()

myData = pd.DataFrame([fake_data() for _ in range(100)], columns = ['Name', 'Gender', 'Address', 'Nationality', 'Account_Type', 'Age','Education', 'Employment', 'Salary', 'Employer_Stability', 'Customer_Loyalty', 'Balance', 'Residential_Status'])
myData.index.names = ['Customer_ID']

tt = time() - t0


print '1 Million rows of data generated in {} seconds'.format(round(tt, 3))



#myData.to_csv('../datasets/dummyTrain.csv')
 
```

```python
# convert the engineered values into real customer details

#myData['Service_Level'] = np.where((myData['Service_Level'] == 0), 'Silver', (np.where((myData['Service_Level'] == 1), 'Gold', 'Platinum')))

myData['Employment'] = np.where((myData['Employment']== 0), 'Student', (np.where((myData['Employment'] == 1), 'Contract', 'Permanent')))

myData['Education'] = np.where((myData['Education']== 0), 'Highschool and below',  'Tertiary and above')

myData['Employer_Stability'] = np.where((myData['Employer_Stability']== 0), 'Unstable',  'Stable')

myData['Residential_Status'] = np.where((myData['Residential_Status'] == 0), 'Rented',  'Owned')

#myData['Customer_Loyalty'] = np.where((myData['Customer_Loyalty'] == 0), '0 - 2 years',  '3 - 10 years')

myData['Age'] = np.where((myData['Age']== 0), '60 +', (np.where((myData['Age'] == 1), '36 - 59', '18 - 35')))

# myData['Gender'] = np.where((myData['Gender'] == 'Male'), 0,  1)

# myData['Account_Type'] = np.where((myData['Account_Type'] == 'Current'), 0,  1)

myData['Salary'] = np.where((myData['Salary']== 0), np.random.randint(150, 1001), (np.where((myData['Salary'] == 1), np.random.randint(1001, 10001), np.random.randint(10001, 100000))))

myData['Balance'] = np.where((myData['Balance']== 0), np.random.randint(20, 2501), (np.where((myData['Balance'] == 1), np.random.randint(2501, 10001), np.random.randint(10001, 500000))))

# np.where(('condition1'<= 6), 0, (np.where(('condition2' > 6), 1, 2)))
```
